## 4. Create Pseudocode

After understanding the task and manually checking the logic, pseudocode was written to plan the structure of the program.

### Planned Functions
The program needed the following functions:

- `load_transactions()`
- `calculate_balance(df)`
- `add_transaction(df)`
- `view_all_transactions(df)`
- `save_transactions(df)`

### Pseudocode

```text
START PROGRAM

IMPORT pandas and pyinputplus
SET FILE_NAME = "budget.csv"
SET COLUMNS = ["date", "description", "type", "amount"]

FUNCTION load_transactions(filename)
    TRY
        read filename into DataFrame
        return DataFrame
    EXCEPT FileNotFoundError
        return empty DataFrame with required columns
END FUNCTION

FUNCTION calculate_balance(df)
    find total income from rows where type is "income"
    find total expense from rows where type is "expense"
    return income minus expense
END FUNCTION

FUNCTION add_transaction(df)
    ask user for date
    ask user for description
    ask user for type (income or expense)
    ask user for amount
    create a new row with these values
    add row to DataFrame
    return updated DataFrame
END FUNCTION

FUNCTION view_all_transactions(df)
    IF DataFrame is empty
        print "No transactions yet."
    ELSE
        print all transactions
END FUNCTION

FUNCTION save_transactions(df, filename)
    save DataFrame to CSV file
END FUNCTION

CREATE test DataFrame
ASSERT balance of test DataFrame is correct
ASSERT balance of empty DataFrame is 0

LOAD transactions from budget.csv or create empty DataFrame

WHILE True
    display menu
    ask user for menu choice

    IF choice is 1
        add transaction
    ELSE IF choice is 2
        view all transactions
    ELSE IF choice is 3
        calculate and display balance
    ELSE IF choice is 4
        save to CSV
        print goodbye
        break loop
END WHILE

END PROGRAM
```
