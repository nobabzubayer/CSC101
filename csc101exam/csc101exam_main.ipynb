```python
import pandas as pd
import pyinputplus as pyip

FILE_NAME = "budget.csv"
COLUMNS = ["date", "description", "type", "amount"]

def load_transactions(filename=FILE_NAME):
    try:
        df = pd.read_csv(filename)
        return df
    except FileNotFoundError:
        return pd.DataFrame(columns=COLUMNS)

def calculate_balance(df):
    income_total = df.loc[df["type"] == "income", "amount"].sum()
    expense_total = df.loc[df["type"] == "expense", "amount"].sum()
    return float(income_total - expense_total)

def add_transaction(df):
    date_value = pyip.inputStr(prompt="Enter date (YYYY-MM-DD): ")
    description = pyip.inputStr(prompt="Enter description: ")
    transaction_type = pyip.inputChoice(["income", "expense"], prompt="Enter type: ")
    amount = pyip.inputFloat(prompt="Enter amount: ", min=0.01)

    new_row = {
        "date": date_value,
        "description": description,
        "type": transaction_type,
        "amount": amount,
    }

    return pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)

def view_all_transactions(df):
    if df.empty:
        print("No transactions yet.")
    else:
        print("\nAll Transactions:")
        print(df.to_string(index=False))

def save_transactions(df, filename=FILE_NAME):
    df.to_csv(filename, index=False)

# Test data for assert statements
test_df = pd.DataFrame([
    {"date": "2026-06-01", "description": "Salary", "type": "income", "amount": 2000.00},
    {"date": "2026-06-02", "description": "Rent", "type": "expense", "amount": 750.00},
    {"date": "2026-06-03", "description": "Food", "type": "expense", "amount": 150.00},
])

empty_df = pd.DataFrame(columns=COLUMNS)

assert calculate_balance(test_df) == 1100.0, "Balance should be 1100.0"
assert calculate_balance(empty_df) == 0.0, "Empty budget should have 0 balance"

transactions = load_transactions()

while True:
    print("\n=== Personal Budget Tracker ===")
    print("1. Add Transaction")
    print("2. View All Transactions")
    print("3. View Current Balance")
    print("4. Save & Exit")

    choice = pyip.inputChoice(["1", "2", "3", "4"], prompt="Choose an option: ")

    if choice == "1":
        transactions = add_transaction(transactions)
        print("Transaction added successfully.")

    elif choice == "2":
        view_all_transactions(transactions)

    elif choice == "3":
        balance = calculate_balance(transactions)
        print(f"Current Balance: ${balance:.2f}")

    elif choice == "4":
        save_transactions(transactions)
        print("Transactions saved to budget.csv")
        print("Goodbye!")
        break
```
