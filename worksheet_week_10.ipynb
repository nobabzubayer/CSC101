1.1 Problem 1: Tip Calculator
1.1.1 The Scenario
You work at a Perth restaurant. The manager wants a simple tool where staff can enter:

The bill amount (e.g., $45.80)
The tip percentage (e.g., 15%)
And get back:

The tip amount
The total amount to charge
1.1.2 What You Must Submit
STEP 1: Understand the Problem

In your own words, what is this program supposed to do? (2-3 sentences)

Answer : This program should let restaurant staff type in a bill amount and a tip percentage, then calculate the tip and the final total the customer needs to pay. It is basically a simple tip calculator that helps staff work out both values quickly and accurately for use at the restaurant

STEP 2: Identify Inputs and Outputs

INPUTS:

[What does the user provide?]
[What type of data? Numbers? Text?]
OUTPUTS:

[What does the program show back?]
[What format?]

Answer : 



INPUTS:

The bill amount

The tip percentage

Both inputs are numeric values

OUTPUTS:

The tip amount

The total amount to charge

Both outputs are displayed as dollar amounts (currency format)


STEP 3: Work It By Hand

Show me you can do this with a calculator. Use this example:

Bill amount: $45.80
Tip percentage: 15%
Write the steps:

[First calculation I would do]
[Second calculation]
[Final step]
[What I would tell the staff member]

Answer:  

Example: Bill amount:  $45.80, Tip percentage: 15%

First calculation I would do:
Convert 15% to a decimal: 15 ÷ 100 = 0.15

Second calculation:
 
Calculate tip amount: $45.80 × 0.15 =  /$6.87 

Final step:
Add tip to bill for total: $45.80 + $6.87 = /$52.67

What I would tell the staff member:
"For the  $45.80  bill with 15% tip, the tip amount is /$6.87 and the total to charge is /$52.67." 



1.2 Problem 2: Late Fee Calculator
1.2.1 The Scenario
You work for a library in Perth. Books are due back in 14 days. For every day late, there’s a /$0.50 fine (maximum $10.00).

The librarian wants a program where they enter:

How many days the book is overdue
And get back:

The fine amount (or “Maximum fine reached” if it’s over $10)
1.2.2 What You Must Submit
STEP 1: Understand the Problem

In your own words, what is this program supposed to do? (2-3 sentences)

Answer:

This program should ask the librarian to enter how many days a book is overdue, then calculate the late fee based on $0.50 for each overdue day. If the fine goes above $10.00, the program should stop at the maximum and show that the maximum fine has been reached.

STEP 2: Identify Inputs and Outputs

INPUTS:

[What information does the librarian provide?]
OUTPUTS:

[What does the program calculate and show?]
[Any special cases?]

Answer:

INPUTS:

Number of days overdue

OUTPUTS:

Fine amount

Displayed in dollars

Special case: if the fine is more than $10.00, show “Maximum fine reached”


STEP 3: Work It By Hand

Show me you understand the logic. Calculate these three cases manually:

Case A: Book is 5 days late

My calculation: 5 × 0.50 = $2.50
Fine: $2.50


Case B: Book is 25 days late

My calculation: 25 × $0.50 = /$12.50 , but the fine cannot go over /$10.00
Fine: $10.00 (Maximum fine reached)
Special note: [Why is this different?]

Special note: This is different because the calculated fine is above the maximum limit, so the program stops at $10.00 and shows the message instead.


Case C: Book is 0 days late

Fine :  $0.00
