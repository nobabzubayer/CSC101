**AI Response:
I can help with that. Here is a simple Python example using pandas:

1. Import the CSV file.
2. Clean the quantity column by removing the dollar sign and converting it to integers.
3. Group the data by product.
4. Calculate total revenue for each product.

```python
import pandas as pd

df = pd.read_csv("sales_data.csv")

df["Quantity"] = df["Quantity"].replace(r"[\$,]", "", regex=True).astype(int)
df["Unit_Price"] = df["Unit_Price"].replace(r"[\$,]", "", regex=True).astype(float)
df["Total_Sale"] = df["Total_Sale"].replace(r"[\$,]", "", regex=True).astype(float)

top_revenue = df.groupby("Product")["Total_Sale"].sum().sort_values(ascending=False).head(5)
top_units = df.groupby("Product")["Quantity"].sum().sort_values(ascending=False).head(5)

print("Top 5 products by revenue:")
print(top_revenue)

print("\nTop 5 products by units sold:")
print(top_units)
```
